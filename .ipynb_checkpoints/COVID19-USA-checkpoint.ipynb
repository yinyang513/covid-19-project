```python
import pandas as pd
```

```python
us = pd.read_csv('./csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv')
usa_death = pd.read_csv('./csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_US.csv')
```

```python
 #confirmed cases in USA
us2 = us.groupby(['Province_State'],as_index=False).sum()
us2.set_index('Province_State').drop(columns = ['UID','code3','FIPS','Lat','Long_']).T.plot(figsize=(30,30),logy=True)
```
