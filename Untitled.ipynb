```python
import pandas as pd
import numpy as np
import altair as alt
from sklearn.linear_model import LinearRegression

TS_PATH = './csse_covid_19_data/csse_covid_19_time_series/'

def read_ts(TS_PATH):
    ts_confirmed = (pd.read_csv(TS_PATH + 'time_series_covid19_confirmed_global.csv',
                               dtype={'Province/State': str, 'Country/Region': str}, 
                               index_col=['Country/Region', 'Province/State', 'Lat', 'Long'])
                    .rename(columns={'Cases': 'Number'}))
    ts_deaths = pd.read_csv(TS_PATH + 'time_series_covid19_deaths_global.csv',
        dtype={'Province/State': str, 'Country/Region': str}, index_col=['Country/Region', 'Province/State', 'Lat', 'Long']
    ).rename(columns={'Cases': 'Number'})
    ts_recovered = pd.read_csv(TS_PATH + 'time_series_covid19_recovered_global.csv',
        dtype={'Province/State': str, 'Country/Region': str}, index_col=['Country/Region', 'Province/State', 'Lat', 'Long']
    ).rename(columns={'Cases': 'Number'})
    # Add "Unresolved" type
    ts_unresolved = ts_confirmed - ts_deaths - ts_recovered
    # combine
    ts_total = pd.concat(
        [
            ts_confirmed, ts_deaths, ts_recovered, ts_unresolved
        ], 
        keys=[
            'Confirmed', 'Deaths', 'Recovered', 'Unresolved',
            'Mortality (Confirmed)', 'Mortality (Resolved)'
            ], 
        names=['Cases']
    ).reset_index()

    ts_total = ts_total.melt(
        id_vars=['Cases', 'Province/State', 'Country/Region', 'Lat', 'Long'], var_name='Date', value_name='Number'
    )
    ts_total = ts_total[~ts_total['Province/State'].fillna('').str.contains(',') | (ts_total['Country/Region'] != 'US')]
    ts_total.loc[:, 'County'] = ts_total['Province/State'].fillna('').apply(lambda x: None if ',' not in x else x.split(',')[0].strip())
    ts_total.loc[:, 'Province/State'] = ts_total['Province/State'].fillna('').apply(lambda x: x if ',' not in x else x.split(',')[1].strip())

    ts_total['Date'] = pd.to_datetime(ts_total['Date'], format='%m/%d/%y', utc=True).dt.to_period('D').astype(str)

    ts_total['Country/Region'] = ts_total['Country/Region'].replace('US', 'United States')
    ts_total = ts_total[ts_total.Number.notnull()]

    return ts_total
```

```python
# functions

def _alt_standard_layout(chart):
    return chart.\
        configure_axis(
            labelFontSize=14,
            titleFontSize=16
        ).\
        configure_legend(
            labelFontSize=14,
            titleFontSize=16
        ).\
        configure_title(
            fontSize=20
        )
```

```python
def get_country(ts_total):
    ts_country = ts_total.\
        groupby(['Country/Region', 'Date', 'Cases']).\
        Number.sum().\
        reset_index()
    return ts_country

ts_country = get_country(ts_total)
```

```python
def get_country_map(ts_total):
    countries = sorted(set(ts_total['Country/Region']))
    country_region_map = dict(ts_total[['Country/Region', 'Province/State']].drop_duplicates())#.set_index('Country/Region').to_dict()
    return countries, country_region_map

countries, country_region_map = get_country_map(ts_total)
```

```python
def get_mortality(ts_to_plot):
    t = ts_to_plot.set_index(['Cases', 'Date']).Number
    t = pd.concat([
        t.loc['Deaths']/t.loc['Confirmed'],
        t.loc['Deaths']/(t.loc['Deaths'] + t.loc['Recovered'])
        ],
        keys=['Confirmed Cases', 'Resolved Cases'],
        names=['Mortality']
    )

    return t.reset_index()

mortality_ts = get_mortality(ts_to_plot)
```

```python
def plot_ahead(ts_to_plot, case='Confirmed', log_scale=False, country=country):

    def _leave_last_x_out(X, y, leave_out=1):
        if not isinstance(X, pd.DataFrame):
            X = pd.DataFrame(X)
        lr = LinearRegression()
        lr.fit(X.iloc[:-leave_out], y.iloc[:-leave_out])
        pred = lr.predict(X.iloc[-1:])
        return pred

    def _step_ahead_fc(X, y, n_ahead=1):
        if not isinstance(X, pd.DataFrame):
            X = pd.DataFrame(X)

        lr = LinearRegression()
        lr.fit(X, y)

        next_date = pd.Timestamp(X.max().iloc[0]) + pd.DateOffset(n=n_ahead)
        pred = lr.predict(pd.to_numeric(pd.Series(next_date)).to_frame())

        next_dt_str = str(next_date.date())
        return pd.Series(pred, index=[next_dt_str])

    def get_step_ahead_fc(ts, case, window=7, last_x_days=7): 

        t = ts[ts.Cases == case].set_index('Date')['Number'].sort_index().iloc[-(last_x_days + window):].copy()
        t = t[t > 0]

        t_insample = t.rolling(window=window+1, min_periods=window+1).apply(
            lambda x: np.exp(_leave_last_x_out(pd.to_numeric(pd.to_datetime(x.index, format='%Y-%m-%d')), np.log(x), leave_out=1)), raw=False
        )

        t_1 = np.exp(_step_ahead_fc(pd.to_numeric(pd.to_datetime(t.iloc[-window:].index, format='%Y-%m-%d')), np.log(t.iloc[-window:]), n_ahead=1))
        t_2 = np.exp(_step_ahead_fc(pd.to_numeric(pd.to_datetime(t.iloc[-window:].index, format='%Y-%m-%d')), np.log(t.iloc[-window:]), n_ahead=2))
        t_3 = np.exp(_step_ahead_fc(pd.to_numeric(pd.to_datetime(t.iloc[-window:].index, format='%Y-%m-%d')), np.log(t.iloc[-window:]), n_ahead=3))

        t_fc = pd.concat(
            [
                t_insample.append(t_1),
                t_2, t_3
            ], keys=['1-Day Ahead', '2-Day Ahead', '3-Day Ahead'], names=['Cases']
        )

        return t_fc.rename('Number').reset_index().rename(columns={'level_1': 'Date'}).dropna()
        
    t_fc = get_step_ahead_fc(ts_to_plot, case=case, last_x_days=14)

    if log_scale:
        y = alt.Y('Number:Q', scale=alt.Scale(type='log'), axis=alt.Axis(format=',.0f'))
    else:
        y = alt.Y('Number:Q', axis=alt.Axis(format=',.0f'))

    actual = alt.Chart(ts_to_plot[ts_to_plot.Cases == case].iloc[-14:, :]).\
        mark_line(point=True).\
        encode(
            x=alt.X('utcyearmonthdate(Date):T', title='Date'),
            y=y,
            color='Cases:N',
            tooltip=[alt.Tooltip('utcyearmonthdate(Date):T', title='Date'), alt.Tooltip('Number:Q', format=',.0f'), 'Cases:N']
        ).\
        properties(
           height=400,
            title=f'3-Day Ahead Prediction ({country})'
        )

    fc = alt.Chart(t_fc).\
        mark_point().\
        encode(
            x=alt.X('utcyearmonthdate(Date):T', title='Date'),
            y=y,
            color='Cases:N',
            tooltip=[alt.Tooltip('utcyearmonthdate(Date):T', title='Date'), alt.Tooltip('Number:Q', format=',.0f'), alt.Tooltip('Cases:N', title='Forecast')]
        )

    return actual + fc

case_to_extrapolate = st.selectbox('Choose what to extrapolate:', ['Confirmed', 'Deaths', 'Recovered', 'Unresolved'])
fc_plot_ph = st.empty() 
log_fc_scale = st.checkbox('Log Forecast Scale')

fc_plot = plot_ahead(ts_to_plot, case=case_to_extrapolate, log_scale=log_fc_scale)

fc_plot_ph.altair_chart(_alt_standard_layout(fc_plot), use_container_width=True)    
```
